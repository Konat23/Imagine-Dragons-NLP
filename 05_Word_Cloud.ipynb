## WordCloud

```python
import pandas as pd
df = pd.read_csv("songs_clean_advance.tsv", sep="\t", encoding="utf-8")
album = pd.read_csv("songs_sentiment.tsv", sep="\t", encoding="utf-8")
SW = set(['oh','oh oh','yeah','hey','la','na','da','oh oh oh','oh oh oh oh','oh oh oh oh oh','oh oh oh oh oh oh','oh oh oh oh oh oh oh','oh oh oh oh oh oh oh oh','oh oh oh oh oh oh oh oh oh'])
df.head()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Agrupar por album y concatenar todas las letras
lyrics = df.groupby("album")["lyric"].apply(" ".join).reset_index()
```

### Word Cloud all albums

```python
# Crear una nube de palabras por album
for i, row in lyrics.iterrows():
    WC = WordCloud(width=400, height=400, background_color="white", collocations=False,stopwords=SW)
    wc = WC.generate(row["lyric"])
    print(WC.words_)
    plt.figure(figsize=(10, 5))
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.title(row["album"])
    plt.show()
```

```python
wordcloud = WordCloud(width = 800, height = 800, 
                    background_color ='white',
                    min_font_size = 10).generate(' '.join(df['lyrics']))
```

### Word cloud Most positive and negative album

#### Generate wordcloud

```python
most_positive = album.sort_values("positive", ascending=False).iloc[0]
most_positive = lyrics[lyrics["album"] == most_positive["album"]]

most_negative = album.sort_values("negative", ascending=False).iloc[0]
most_negative = lyrics[lyrics["album"] == most_negative["album"]]
```

```python
most_positive["lyric"].values[0]
```

```python
# Wordcloud de las letras del album ms positivo
WC = WordCloud(width=800, height=800, background_color=None,mode="RGBA", stopwords=SW, random_state=42)
wc = WC.generate(most_positive["lyric"].values[0])
WC.to_file("outputs/positive_wordcloud.png")
plt.figure(figsize=(5, 5))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.title(f"Most positive album: {most_positive['album'].values[0]}")
plt.show()
```

```python
# Wordcloud de las letras del album ms negativo
WC = WordCloud(width=800, height=800, background_color=None,mode="RGBA", stopwords=SW, random_state=42)
wc = WC.generate(most_negative["lyric"].values[0])
WC.to_file("outputs/negative_wordcloud.png")
plt.figure(figsize=(5, 5))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.title(f"Most negative album: {most_negative['album'].values[0]}")
plt.show()
```

#### Edit wordcloud

```python
from PIL import Image
import numpy as np

# Cargar las imgenes
cover_p = Image.open("covers/Imagine Dragons - Mercury - Acts 1 & 2 album cover.jpg")
wordcloud_p = Image.open("outputs/positive_wordcloud.png")

# Escalar la imagen de la portada a 800x800
cover_p = cover_p.resize((800, 800))

# Convertir la imagen del wordcloud a un arreglo numpy
wordcloud_array = np.array(wordcloud_p)

# Crear una mscara inversa
inverse_mask = np.where(wordcloud_array.sum(axis=2) == 0, 255, 0).astype(np.uint8)
inverse_mask = Image.fromarray(inverse_mask)
# Volver a invertir la mscara
inverse_mask = Image.eval(inverse_mask, lambda x: 255 - x)

# Aplicar la mscara inversa a la imagen de la portada
cover_p.paste((255, 255, 255), mask=inverse_mask)

# Guardar la imagen resultante
cover_p.save("cover_with_wordcloud_cutout.png")
```
